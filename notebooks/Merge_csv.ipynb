# Rename KAUH recordings

```python
for filename in os.listdir('D:\\lung sounds\\KAUH\\recordings'):
    print(filename)
    new_filename = int(filename.split('.')[0])
    if new_filename < 10:
        new_filename = '00' + str(new_filename) + '.wav'
    elif new_filename < 100:
        new_filename = '0' + str(new_filename) + '.wav'
    else:
        new_filename = str(new_filename) + '.wav'
    print(new_filename)
    
    os.rename(os.path.join('D:\\lung sounds\\KAUH\\recordings', filename),
             os.path.join('D:\\lung sounds\\KAUH\\recordings', new_filename))
```
